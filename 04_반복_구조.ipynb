### ë¬¸ì œ: ì „ ì§ì›ì˜ ì›”ê¸‰ ì´ì•¡ êµ¬í•˜ê¸°
*   ì§ê¸‰ì´ 2ê¸‰ ì´ìƒì´ë©´ ì›”ê¸‰ì´ ë³¸ë´‰+ìƒì—¬ê¸ˆ(ë³¸ë´‰ì˜30%)
*   ì§ê¸‰ì´ 2ê¸‰ ë¯¸ë§Œì´ë©´ ì›”ê¸‰ì´ ë³¸ë´‰+ìƒì—¬ê¸ˆ(ë³¸ë´‰ì˜20%)
*   ë³€ìˆ˜ employee_num ì§ì›3ëª…ì˜ ì›”ê¸‰ ê³„ì‚°
*   ë³€ìˆ˜ total í•©ê³„ ì¶œë ¥
*   ë³€ìˆ˜ rank ì§ê¸‰ ì…ë ¥ ë°›ê³ 
*   ë³€ìˆ˜ basic_salary ë³¸ë´‰ ì…ë ¥ ë°›ê³ 
*   ë³€ìˆ˜ bonus ìƒì—¬ê¸ˆ
*   ë³€ìˆ˜ salary ì›”ê¸‰







```python
employee_num = 3
total = 0

for i in range(employee_num): # 0,1,2
  rank = int(input('ì§ê¸‰: '))
  basic_salary = int(input('ë³¸ë´‰: '))
  if rank >= 2:
    bonus = basic_salary * 0.3
  else:
    bonus = basic_salary * 0.2
  salary = basic_salary + bonus
  total += salary

print(f'ì „ ì§ì› ì›”ê¸‰ ì´ì•¡: {total}')
```

### ë¬¸ì œ: ê²½ì§„ëŒ€íšŒ í‰ê°€ ì ìˆ˜ êµ¬í•˜ê¸°
*   ì‹¬ì‚¬ìœ„ì› judge_num ì´ 5ëª…ì´ ì‹¬ì‚¬
*   score ê° ì ìˆ˜ë¥¼ ì…ë ¥ ë°›ê³ 
*   ì ìˆ˜ ì´ í•©ê³„ sumì˜ í‰ê· ì„ ì¶œë ¥
*   ìµœê³  ì ìˆ˜ëŠ” ë¹¼ê³  í•©ê³„ë¥¼ êµ¬í•´ì•¼ í•¨ max_score
*   í˜„ì¬ ì ìˆ˜ê°€ ìµœê³  ì ìˆ˜ë³´ë‹¤ í¬ë©´, í˜„ì¬ ì ìˆ˜ë¥¼ max_scoreì— ì €ì¥í•˜ê³  sumì—ì„œ ë¹¼ì¤˜ì•¼ í•©ë‹ˆë‹¤.


```python
judge_num = 5
sum = 0
max_score = 0

for i in range(judge_num): # 0,1,2,3,4
  score = int(input('ì ìˆ˜: ')) # 100, 80, 70, 90, 100
  sum += score # 100+80+70+90+100
  if score > max_score:
    max_score = score # 100
sum -= max_score
print(f'ì‘í’ˆ í‰ê°€ ì ìˆ˜: {sum/(judge_num-1)}')
```

### ì ì • ì‹¬ë°•ìˆ˜ êµ¬í•˜ê¸°

*   ìµœëŒ€ ìš´ë™ ê°•ë„ë¥¼ 40%, 50%, 60%, 70%, 80%ì— í•´ë‹¹í•˜ëŠ” ëª©í‘œ ì‹¬ë°•ìˆ˜ë¥¼ ê³„ì‚°í•˜ëŠ” ì‹
*   ë³€ìˆ˜ intensity ìš´ë™ ê°•ë„ëŠ” 40/100 ê³„ì‚°
*   ë³€ìˆ˜ resting_HR ì•ˆì •ì‹œ ì‹¬ë°•ìˆ˜
*   ë³€ìˆ˜ target_HR ëª©í‘œ ì‹¬ë°•

*   ë‚˜ì´age ì…ë ¥ ë°›ê³

*   ê°•ë„ ëª‡ % ì¼ë•Œ ëª©í‘œ ì‹¬ë°•ìˆ˜ê°€ ì–¼ë§ˆë‹¤ ë¼ê³  ì¶œë ¥



```python
age = int(input('ë‚˜ì´: '))
resting_HR = int(input('ì•ˆì •ì‹œ ì‹¬ë°•ìˆ˜: '))

for i in range(40,81,10): # 40,50,60,70,80
  intensity = i/100
  target_HR = (((220-age)-resting_HR)*intensity)+resting_HR
  print(f'ìš´ë™ ê°•ë„: {i}% ì¼ë•Œ ëª©í‘œ ì‹¬ë°•ìˆ˜: {target_HR} bpm')
```

## 5. [ì‹¤ìŠµ ì˜ˆì œ]

### ğŸ¯ ìˆ«ì ë§íˆê¸° ê²Œì„
1. **ì»´í“¨í„°ê°€ ìƒê°í•œ ìˆ˜ ë§íˆê¸°**:
    * `random.randint(1, 100)`ìœ¼ë¡œ ì •ë‹µì„ ìƒì„±í•©ë‹ˆë‹¤.
    * ì‚¬ìš©ìê°€ ì •ë‹µì„ ë§í ë•Œê¹Œì§€ `while` ë¬¸ìœ¼ë¡œ ì…ë ¥ì„ ë°˜ë³µí•©ë‹ˆë‹¤.
    * ì…ë ¥ê°’ì´ í¬ë©´ "ë„ˆë¬´ ë†’ì•„ìš”", ì‘ìœ¼ë©´ "ë„ˆë¬´ ë‚®ì•„ìš”"ë¥¼ ì¶œë ¥í•˜ì—¬ ë²”ìœ„ë¥¼ ì¢í˜€ê°‘ë‹ˆë‹¤.
2. **ì‚¬ìš©ìê°€ ìƒê°í•œ ìˆ˜ ë§íˆê¸°**:
    * ì‚¬ìš©ìê°€ ìˆ«ìë¥¼ í•˜ë‚˜ ì •í•˜ë©´ ì»´í“¨í„°ê°€ `low`ì™€ `high` ë³€ìˆ˜ë¥¼ ì¡°ì •í•˜ë©° ì •ë‹µì„ ì°¾ì•„ê°€ëŠ” ë¡œì§ì…ë‹ˆë‹¤.

```python
# 1. ì»´í“¨í„°ê°€ ìƒê°í•œ ìˆ«ì ë§ì¶”ê¸°
import random

print('1 ë¶€í„° 100 ê¹Œì§€ì˜ ìˆ«ì í•˜ë‚˜ë¥¼ ì •í–ˆìŠµë‹ˆë‹¤.')
print('ì´ ìˆ«ìë¥¼ ë§ì¶°ë³´ì„¸ìš”!')

answer = random.randint(1,100) # 1~100ì‚¬ì´ ìˆ«ì í•˜ë‚˜ ëœë¤ìœ¼ë¡œ ì €ì¥
times = 0 # ë„ì „ íšŸìˆ˜
# print(f'ì»´í“¨í„°ê°€ ìƒì„±í•œ ìˆ«ì: {answer}')

# while ë¬¸ìœ¼ë¡œ ë°˜ë³µí•˜ëŠ”ë°, ì¦‰ ë„ì „ì´ ê³„ì†
while True:
  user = int(input('1~100 ì‚¬ì´ì˜ ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”!'))
  times += 1 # ë„ì „ íšŸìˆ˜ 1 ì¦ê°€
# userê°€ ì…ë ¥í•œ ìˆ«ìê°€ ë§ìœ¼ë©´ ë°˜ë³µë¬¸ ì¢…ë£Œê°€ ë˜ê³ 
  if user == answer:
    break
# userê°€ ì…ë ¥í•œ ìˆ«ì í¬ë©´ í¬ë‹¤ê³  ì¶œë ¥
  elif user > answer:
    print(f'{user}ë³´ë‹¤ ì‘ìŠµë‹ˆë‹¤')
# userê°€ ì…ë ¥í•œ ìˆ«ì ì‘ìœ¼ë©´ ì‘ë‹¤ê³  ì¶œë ¥
  else:
    print(f'{user}ë³´ë‹¤ í½ë‹ˆë‹¤')
# ìµœì¢… ë„ì „ ëª‡ íšŒë§Œì— ì •ë‹µì„ ë§ì·„ëŠ”ì§€ ì¶œë ¥
print(f'ì •ë‹µì…ë‹ˆë‹¤! {times}íšŒ ë§Œì— ë§ì·„ìŠµë‹ˆë‹¤.')
```

```python
# 2.ì‚¬ìš©ìê°€ ìƒê°í•œ ìˆ˜ ë§íˆê¸°
print('1~100 ì‚¬ì´ ìˆ«ì í•˜ë‚˜ë¥¼ ì •í•˜ì„¸ìš”')
print('ì´ ìˆ«ìë¥¼ ë§í˜€ë³´ê² ìŠµë‹ˆë‹¤!')

low = 1
high = 100
times = 0

while True:
  print(f'low: {low}, high: {high}')
  num = (low+high)//2  # ì»´í“¨í„°ê°€ ì œì‹œí•œ ìˆ«ì
  answer = input(f'{num}ì¸ê°€ìš”? ok / up /down') # ì‚¬ìš©ìê°€ ë‹µë³€ì„ í•´ì¤Œ
  times += 1

  if answer == 'ok':
    break
  elif answer == 'up': # numë³´ë‹¤ í¬ë‹¤ê³  í•´ì„œ lowê°€ num+1
    low = num+1
  elif answer == 'down': # numë³´ë‹¤ ì‘ë‹¤ê³  í•´ì„œ highê°€ num-1
    high = num-1
  else:
    print('ì˜ëª» ì…ë ¥í–ˆìŠµë‹ˆë‹¤.')

print(f'{times}íšŒ ë§Œì— ë§í˜”ìŠµë‹ˆë‹¤.')
```
