## 2. í´ë˜ìŠ¤ ìƒì„± ë° ì‚¬ìš©

* **ë©”ì„œë“œ ì˜¤ë²„ë¼ì´ë”©(Overriding)**: ë¶€ëª¨ í´ë˜ìŠ¤ë¡œë¶€í„° ë¬¼ë ¤ë°›ì€ ë©”ì„œë“œë¥¼ ìì‹ í´ë˜ìŠ¤ì—ì„œ ìš©ë„ì— ë§ê²Œ ìƒˆë¡­ê²Œ ì¬ì •ì˜í•˜ì—¬ ì‚¬ìš©í•˜ëŠ” ê²ƒì„ ë§í•©ë‹ˆë‹¤.

```python
# ë¶€ëª¨ í´ë˜ìŠ¤
class Student:
 def __init__(self, name, computer):
    self.name = name
    self.computer = computer

```

```python
# ìì‹ í´ë˜ìŠ¤
class Science(Student):
  def __init__(self, name, computer, science):
    super().__init__(name, computer)
    self.science = science

  def set_science(self, science):
    self.science = science

  def get_average(self):
    return (self.science + self.computer) /2
```

```python
#ìì‹ í´ë˜ìŠ¤
class Social(Student):
  def __init__(self, name, computer, social):
    super().__init__(name, computer)
    self.social = social

  def set_social(self, social):
    self.social = social

  def get_avarage(self):
    return (self. social + self.computer) /2
```

```python
#ë¶€ëª¨ í´ë˜ìŠ¤
class Figure:
  def __init__(self, area, perimeter):
    self.area =area
    self.perimeter = perimeter

  def get_area(self):
    return self.area

  def get_perimeter(self):
    return self.perimeter
```

```python
#ìì‹ í´ë˜ìŠ¤
class Circle(Figure):
  def __init__ (self, radius):
    area = 3.14*radius**2
    perimeter = 2*3.14*radius
    super().__init__(area, perimeter)
    self.radius = radius

  def set_radius(self, radius):
    self.radius = radius
    self.area = 3.14*radius**2
    self.perimeter = 2*3.14*radius

  def get_radius(self):
    return self.radius
```

```python
#ìì‹ í´ë˜ìŠ¤
class Rectangle(Figure):
  def __init__(self, width=10, height = 10):
    area = width*height
    perimeter = 2*width+2*height
    super().__init__(area, perimeter)
    self.width = width
    self.height = height

  def set_width(self, width):
    self.width = width
    self.area = self.width*self.height
    self.perimeter = 2*self.width+2*self.height

  def set_height(self, height):
    self.height = height
    self.area = self.width*self.height
    self.perimeter = 2*self.width+2*self.height

  def get_width(self):
    return self.width

  def get_height(self):
    return self.height

```

```python
f1 = Circle(7)
print(f1.area)
print(f1.perimeter)
print(f1.get_area(), f1.get_perimeter())
f1.set_radius(10)
print(f1.get_area(), f1.get_perimeter())
```

```python
f2 = Rectangle(5,4)
print(f2.get_area(). f2.get_perimeter())
f2.set_width(10)
print(f2.get_area(), f2.get_perimeter())
f2.set_height(5)
print(f2.get_area(), f2.get_perimeter())
```
